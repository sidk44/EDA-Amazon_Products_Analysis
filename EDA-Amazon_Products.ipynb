```python
#Creating important functions for data cleaning
def missing_value(data):
    return data.isnull().sum()

def data_type(data):
    return data.dtypes

def check_duplicate(data):
    return data.duplicated().sum()

print(missing_value(amazon_df))
print('\n-----------------------------\n')
print(amazon_df.shape,'\n')
print(check_duplicate(amazon_df))
print('\n-----------------------------\n')
data_type(amazon_df)
```

```python
amazon_df.dropna(axis=0,subset=['rating_count'])
amazon_df[amazon_df.rating_count.isnull()]
```

```python
#dropping rows with missing info , i.e. NaN
amazon_df.dropna(axis=0,subset=['rating_count'],inplace=True)
print(missing_value(amazon_df))
```

```python
amazon_df['discounted_price'] = amazon_df['discounted_price'].astype(str).str.replace('₹','').str.replace(',','').astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].astype(str).str.replace('₹','').str.replace(',','').astype(float)
amazon_df['discount_percentage'] = (amazon_df['discount_percentage'].astype(str).str.replace('%','').astype(float))/100
amazon_df
```

```python
#more data cleaning
amazon_df['rating'] = amazon_df['rating'].astype(str)
count = amazon_df['rating'].str.contains('\|').sum()
print("Total number of rows with '|' in the 'rating' column: ",count)
amazon_df = amazon_df[amazon_df['rating'].apply(lambda x: '|' not in str(x))]
print("Total number of rows with '|' in the 'rating' column: ",count)

#result shows 0 in both outputs because it was run more than once, so the dataframe gets updated.
```

```python
amazon_df['rating'] = amazon_df['rating'].astype(str).str.replace(',','').astype(float)
amazon_df['rating_count'] = amazon_df['rating_count'].astype(str).str.replace(',','').astype(float)
```

```python
check_datatype(amazon_df)
#great, we just changed their datatype to float , so we can perform operations on them.
```

```python
# Creating the column "rating_weighted"
amazon_df['rating_weighted'] = amazon_df['rating']*amazon_df['rating_count']
```

```python
#Extracting main and sub categories from the category column.
amazon_df['main_category'] = amazon_df['category'].astype(str).str.split('|').str[0]
amazon_df['sub_category'] = amazon_df['category'].astype(str).str.split('|').str[-1]
print(amazon_df.columns,'\n\n',"No. of rows is:",len(amazon_df))
amazon_df.head()
```

```python
#all statistical measures
amazon_df.describe()
```

```python
#correlation 
#positive correlation : one value increases other also increases (we compare columns)
#negative correlation: one value increases other decreases.

corr1 = amazon_df['discounted_price'].corr(amazon_df['rating']) #specific 
corr2 = amazon_df['discount_percentage'].corr(amazon_df['rating'])
print(corr1,corr2)
#all at one command
float_columns = amazon_df.select_dtypes(include=['float64'])
float_columns.corr()
```

```python
#data analysis and findings
import matplotlib.pyplot as plt
# Analyzing the distribution of products by main category
main_category_count = x = amazon_df['main_category'].value_counts()[:30]  #top30(we can say top 10 also, as there are less columns)
y = main_category_count.values
plt.bar(range(len(x)),y)
plt.ylabel('No. of Products')
plt.title('Distribution of Top 30 Products in Main Category')
plt.xticks(range(len(x)),'')
plt.show()
# print(amazon_df['main_category'].value_counts()[:30])
# main_category_count.index
top_main_categories = pd.DataFrame({'Main Category': main_category_count.index, 'Number of Products': main_category_count.values})
print('Top 30 main categories:')
top_main_categories


```

```python
# Analyzing distribution of products by last category
sub_category_count = x = amazon_df['sub_category'].value_counts()[:30]
y = sub_category_count.values
plt.bar(range(len(sub_category_count)), sub_category_count.values)
plt.ylabel('No. of Products')
plt.title(' Distribution of Top 30 Products by Sub Category')
plt.xticks(range(len(sub_category_count)),'')
plt.show()

# Top 30 sub categories 
top_sub_categories = pd.DataFrame({'Sub Category':x.index, 'No. of Products':y})
print("Top 30 sub categories \n")
print(top_sub_categories.to_string(index=False))
# we found some findings in sales from this analysis
```

```python
#Analyze the distribution of customer ratings using a histogram.
plt.hist(amazon_df['rating'],color='orange',bins=20) #bins means no. of bars you see in graph.
plt.xlabel('Rating')
plt.ylabel('No. of Reviews')
plt.title('Distribution of Customer Ratings')
plt.show()
#we can see what rating was given in majority

# Create table with values per cluster
bins = [0, 1, 2, 3, 4, 5] # Define bin edges
amazon_df['cluster'] = pd.cut(amazon_df['rating'], bins=bins, include_lowest=True, labels=['0-1', '1-2', '2-3', '3-4', '4-5'])
table = amazon_df['cluster'].value_counts().reset_index().sort_values('cluster').rename(columns={'count': 'No. of Reviews', 'cluster': 'Cluster'})
print(table)
```

```python
# Calculate the top main categories
#we obtain some important info about popularity of main-categories.
top_m1 = amazon_df.groupby(['main_category'])['rating'].mean().sort_values(ascending=False).head(10).reset_index()
plt.bar(top_m1['main_category'],top_m1['rating'])

plt.xlabel('Main Category -->')
plt.ylabel('Rating -->')
plt.title('Top main_category by Rating')
plt.xticks(rotation=90)
plt.show()
top_m1
#REVISE!!
```

```python
top_s1 = amazon_df.groupby('sub_category')['rating'].mean().sort_values(ascending=False).head(10).reset_index()
plt.bar(top_s1['sub_category'],top_s1['rating'])
plt.xlabel('Sub Category ')
plt.ylabel('Rating')
plt.title('Top Sub Category by Rating')
plt.xticks(rotation=90)
plt.show()
top_s1
```

```python
# Lets find mean discount by each main category
# sort the means in descending order
mean_discount_by_category = amazon_df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)*100
plt.barh(mean_discount_by_category.index,mean_discount_by_category.values)
plt.ylabel('Main Category')
plt.xlabel('Discount %')
plt.title('Mean discount by each Main Category')

plt.show()
print(mean_discount_by_category/100)
#that is only for graph we multiplied 100 for readability
```

```python
mean_discount_by_sub_category = amazon_df.groupby('sub_category')['discount_percentage'].mean().head(15).sort_values()*100
plt.barh(mean_discount_by_sub_category.index,mean_discount_by_sub_category.values)
plt.ylabel('Sub Category')
plt.xlabel('Discount')
plt.title('Discount percentage by sub category')
plt.show()
mean_discount_by_sub_category/100
```

```python
#Analyze the reviews by creating word clouds 
# OR frequency tables of the most common words used in the reviews.
from wordcloud import WordCloud,STOPWORDS
review_txt = ' '.join(amazon_df['review_content'].dropna().values)
#joins all the review content in entire column separated by space ' '.
worldcloud = WordCloud(stopwords=set(list(STOPWORDS) + ['support','using']) , width=699, height=499,background_color='white',min_font_size=5).generate(review_txt)
#first parameter is to neglect some set of words like 'is,an,can' & 'support' etc
plt.imshow(worldcloud)
plt.axis('off')
plt.show()
```
